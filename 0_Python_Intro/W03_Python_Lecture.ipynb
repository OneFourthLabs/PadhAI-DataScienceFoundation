# Functions

```python
def fibonacci(pos):
    a = 1
    b = 1
    for i in range(pos):
        temp = a + b
        a = b
        b = temp
    return temp
```

```python
print(fibonacci(3))
```

```python
print(fibonacci(8), fibonacci(7))
```

```python
for i in range(2, 20):
    ratio = fibonacci(i) / fibonacci(i - 1)
    print(i, ratio)
```

```python
def fibonacci_relative(pos, a, b):
    for i in range(pos):
        temp = a + b
        a = b
        b = temp
    return temp
```

```python
print(fibonacci_relative(3, 34, 55))
```

```python
def fibonacci_ol(pos, a = 1, b = 1):
    for i in range(pos):
        temp = a + b
        a = b
        b = temp
    return temp
```

```python
print(fibonacci_ol(3, 34, 55))
```

```python
print(fibonacci_relative(3))
```

```python
def fibonacci_recursive(n, a = 1, b = 1):
    if n > 1:
        return fibonacci_recursive(n - 1, b, a + b)
    else:
        return a + b
```

```python
print(fibonacci_recursive(3, 34, 55))
```
